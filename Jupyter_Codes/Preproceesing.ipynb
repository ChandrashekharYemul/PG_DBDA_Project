```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.tokenize import sent_tokenize, wordpunct_tokenize
from nltk.corpus import stopwords
```

```python
df = pd.read_csv("SportsDataNf2.csv",encoding='latin-1')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.drop(['ï»¿Artical Id', 'Unnamed: 3','Unnamed: 4','Unnamed: 5'], inplace=True, axis=1)
df.tail()
```

```python
# df.dropna()
df['Artical Data'] = df['Artical Data'].fillna('')
```

```python
# Sentence Tokenization 
sentences = []
for text in df['Artical Data']:
    sentences.extend(sent_tokenize(text))
sentances
```

```python
from nltk.tokenize import wordpunct_tokenize
# wordpunct_tokenize(corpus)
words = []
for word in df['Artical Data']:
    words.extend(wordpunct_tokenize(word))
words
```

```python
total_word = list(words)
len(total_word)
```

```python
from nltk import word_tokenize
lower_word = []
for word in df['Artical Data']:
    lower_word.extend(word_tokenize(word.lower()))
lower_word
```

```python
total_word = list(lower_word)
len(total_word)
```

```python
from nltk.corpus import stopwords
clean_word =[]
words_to_remove = [',', '?', '.', ':', '', '!', '@', '$', '%', '^', '*', '"', '(' , ')', '<', '>', '&', ';', '[', ']', '\\\n','``','\\n','\'\'','\'','\'s']
# words_to_remove = [",", ".", ":", "\\/", "!", "@", "$", "%", "^", "*","(" , ")", "<", ">", "&", ";", "[", "]", "\n","'",""",]

words_to_remove.extend(stopwords.words('english'))
clean_word.extend([word.strip() for word in lower_word if word.strip() not in words_to_remove])
clean_word
```

```python
total_word = list(clean_word)
len(total_word)
```
